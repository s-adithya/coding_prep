```python
def twoSum(nums: List[int], target: int) -> List[int]:
    sort_nums = sorted(nums)
    for i in range(len(sort_nums)//2 +1):
        for j in range(i+1 , len(sort_nums)):
            if sort_nums[i] + sort_nums[j] == target:
                return [i,j]
                
    return []
    
                
nums = [2,4,5,6] 
target  = 11              
```
